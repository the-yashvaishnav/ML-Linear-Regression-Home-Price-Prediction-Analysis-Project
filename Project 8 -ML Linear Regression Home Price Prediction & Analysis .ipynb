## Part-2: Working with Model 

```python
# 1) Create the target data and feature data where target data is price 
```

```python
x=data.drop("Price",axis=1)
y=data["Price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#scaling
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train=pd.DataFrame(scaler.fit_transform(x_train),columns=x.columns)
x_test=pd.DataFrame(scaler.transform(x_test),columns=x.columns)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
data
```

```python
yp=model.predict(x_test)
```

```python
print("coef:-",model.coef_)
print("intercept_:-",model.intercept_)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,root_mean_squared_error,r2_score
```

```python
print("mean_squared_error:-",mean_squared_error(yp,y_test))
print("mean_absolute_error:-",mean_absolute_error(yp,y_test))
print("root_mean_squared_error",root_mean_squared_error(yp,y_test))
print("r2_score",r2_score(yp,y_test)*100)
```
